```python
import pandas as pd
import numpy as np
import yfinance as yf
from matplotlib import pyplot as plt

#取得台積電股價

start = '2016-12-22'
end = '2021-06-04'
df1 = yf.download("00677U.TW", start=start, end=end) #富邦VIX
df2 = yf.download("^VIX", start=start, end=end) #VIX指數

from matplotlib import pyplot as plt

plt.figure(figsize=(20,6))
plt.plot(df1.index, df1['Close'], label='富邦VIX ETF')
plt.plot(df2.index, df2['Close'], label='VIX')

plt.title("富邦VIX ETF v.s. VIX", fontsize=16)
plt.ylabel('價格', fontsize=16)
plt.xlabel("日期", fontsize=16)
plt.grid(axis='both')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=16)
plt.xticks(rotation=45, fontsize=16)
plt.yticks(fontsize=16)
plt.savefig('/content/drive/MyDrive/政大課程/111選擇權評價與應用/期末報告/富邦VIX.png', bbox_inches='tight')
plt.show()
```

```python
start = '2016-12-22'
end = '2021-06-04'
df1 = yf.download("00677U.TW", start=start, end=end) #富邦VIX
df2 = yf.download("^VIX", start=start, end=end) #VIX指數

ER = pd.read_csv("/content/drive/MyDrive/政大課程/111選擇權評價與應用/期末報告/VIXFuturesER.csv").iloc[:,:2].dropna()
ER.columns = ['date', 'ER']
ER['date'] = pd.to_datetime(ER['date'])
ER = ER[(ER['date']>=start)&(ER['date']<=end)]



plt.figure(figsize=(20,6))

plt.plot(df2.index, df2['Close'].pct_change()*100, label='VIX', alpha=0.7)
plt.plot(df1.index, df1['Close'].pct_change()*100, label='富邦VIX ETF', alpha=0.7)
plt.plot(ER["date"], ER['ER'].pct_change()*100, label='VIX futures ER', alpha=0.7)

plt.title("VIX, 富邦VIX ETF, VIX futures ER報酬比較", fontsize=16)
plt.xlabel("date", fontsize=16)
plt.ylabel("報酬(%)", fontsize=16)
plt.grid(axis='both')
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), fontsize=16)
plt.xticks(rotation=45, fontsize=16)
plt.yticks(fontsize=16)
plt.savefig('/content/drive/MyDrive/政大課程/111選擇權評價與應用/期末報告/報酬比較表.png', bbox_inches='tight')
plt.show()
```

```python
start = '2016-12-22'
end = '2021-06-04'
df1 = yf.download("^GSPC", start=start, end=end) #SP500
df2 = yf.download("^VIX", start=start, end=end) #VIX

display(df1)

fig,ax = plt.subplots(figsize=(20,6))

ax.plot(df1.index,  df1['Close'], color='blue')
ax.set_xlabel("Date", fontsize = 14, color='black')
ax.set_ylabel('S&P 500', fontsize=14, color='blue')
ax.yaxis.set_tick_params(labelsize=14)
ax.xaxis.set_tick_params(labelsize=14)

ax2=ax.twinx()
ax2.plot(df2.index,  df2['Close'], color='red')
ax2.set_ylabel("VIX",color="red",fontsize=14)
ax2.yaxis.set_tick_params(labelsize=14)

plt.grid(axis='both')

plt.savefig('/content/drive/MyDrive/政大課程/111選擇權評價與應用/期末報告/VIXSP500.png', bbox_inches='tight')
plt.show()
```
